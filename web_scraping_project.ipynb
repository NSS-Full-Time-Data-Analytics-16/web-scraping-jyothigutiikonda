### b. Now, use what you did for the first title, but extract the job title for all jobs on this page. Store the results in a list.  
### c. Finally, extract the companies, locations, and posting dates for each job. For example, the first job has a company of "Payne, Roberts and Davis", a location of "Stewartbury, AA", and a posting date of "2021-04-08". 
### Ensure that the text that you extract is clean, meaning no extra spaces or other characters at the beginning or end. 

```python
lists = soup.findAll('h2')
jobs_list = []
for item in lists:
    jobs_list.append(item.text.strip())
jobs_list 
```

```python
com  = soup.findAll('h3')
companies = []
for item in com:
    companies.append(item.text.strip())
companies   
```

```python
loc = soup.findAll('p',class_='location')
locations = []

for item in loc:
    locations.append(item.text.strip())
locations
```

```python
date = soup.findAll('time')
posting_date = []
for item in date:
    posting_date.append(item.text.strip())
posting_date
```

### d. Take the lists that you have created and combine them into a pandas DataFrame. 

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Job_title': jobs_list,
    'Company':companies,
    'location':locations,
    'posting_date':posting_date
})
```

```python
def get_descriptions(url):
    response = requests.get(url)
    soup = BS(response.text)
    return soup.find(class_="content").find("p").text
    

```
